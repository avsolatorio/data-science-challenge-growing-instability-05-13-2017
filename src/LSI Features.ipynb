```python
from gensim.models import ldamulticore, lsimodel, tfidfmodel, word2vec
from gensim.corpora import Dictionary, MmCorpus
```

```python
lsn = word2vec.LineSentence(
    'corpus/train_body_data-with_labels_False-retain_special_chars_False.with_test_data.csv'
)
```

```python
tfidf = tfidfmodel.TfidfModel.load(
    lsn.source + '.tfidf'
)

dictionary = tfidfmodel.TfidfModel.load(
    lsn.source + '.dictionary'
)
```

```python
%%time
def _corpus_generator(lsn):
    for l in lsn:
        yield dictionary.doc2bow(l)
```

```python
%%time
corpus_generator = _corpus_generator(lsn)

lsi = lsimodel.LsiModel(
    tfidf[corpus_generator],
    num_topics=300,
    id2word=dictionary.id2token,
)
lsi.save(lsn.source + '.lsi')
```
